```
# Creating local Spark session
import findspark
findspark.init()
from pyspark.sql import SparkSession
import pyspark.sql.functions as func
from pyspark.sql.types import IntegerType
spark = SparkSession.builder.master("local[*]").getOrCreate()
```

```
#read csv file
df = spark.read.format("csv").option("header", "true").load("/content/drive/My Drive/Amazon_Responded_Oct05.csv")
df.head()
```

```
#Keeping relevant columns as specified in the assignment
df_new = df.select('id_str','tweet_created_at','user_verified','favorite_count','retweet_count','text_')
df_new.head(5)
```

#TASK1


Step-1

```
#Removing records where user verified = false
df11 = df_new.where(df_new.user_verified == "True")
df11.count()
```

Step-2

```
#Extracting date in proper format & replacing it in a different variable
df12 = df11.withColumn('tweet_created_new', df11['tweet_created_at'].substr(5,6))
df12.show()#new date column created as tweet_created_new
```

```
#Grouping by new date and counting number of tweets for each date
date_count = df12.groupby("tweet_created_new").count()
date_count.sort('count', ascending=False).show()
#Jan3 has highest number of tweets
```

Step-3

```
df12.dtypes #favorite_count & retweet_count has data type of string which needs to be type casted into an integer
df12 = df12.withColumn("favorite_count", df12["favorite_count"].cast(IntegerType()))
df12 = df12.withColumn("retweet_count", df12["retweet_count"].cast(IntegerType()))
df12.dtypes
```

```
#Creating anew column of total_count as sum of favourite & retweet counts
df13 = df12.withColumn('total_count', df12.favorite_count+ df12.retweet_count)
df13.show()
```

```
#Reporting text  of topp 100 tweets by total_count
date_count = df13.where(df13.tweet_created_new == "Jan 03").sort('total_count', ascending=False).select('text_','total_count').show(n=100)
```

```
#DataCleaning
from pyspark.sql.functions import udf, col, lower, regexp_replace
#Cleaning text by removing punctuations & changing it to lower & adding the clean data to new column text_new
df_clean = df13.where(df13.tweet_created_new == "Jan 03").sort('total_count', ascending=False).select('text_').select('text_', (lower(regexp_replace('text_', "[^a-zA-Z\\s]", "")).alias('text_new'))).limit(100)
df_clean.withColumn('text_new', regexp_replace('text_new', "http\S+", ""))
df_clean.select('text_new').show()
```

```
#Applying map reduce functionaly to get the count frequency of 100 tweets
df_final = df_clean.select("text_new").rdd.flatMap(lambda x: x[0].split(' ')).map(lambda x: (x, 1)).reduceByKey(lambda x,y: x+y)
df_final.take(100)
```

```
#Converting RDD to DataFrame & Writing answer to csv
Answer = df_final.toDF()
Final_Answer = Answer.withColumnRenamed("_1","Word").withColumnRenamed("_2","Frequency")
Final_Answer.show(637)
Final_Answer.write.format("csv").save('/content/drive/My Drive/Amazon_Output.csv', header=True)
```
